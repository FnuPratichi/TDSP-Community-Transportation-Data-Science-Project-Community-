> Answer here:

- Bias in datasets often arises when data availability is skewed, particularly when more data is collected from well-resourced communities compared to under-resourced ones. This can lead to several ethical issues, such as:

- Representation Bias: Well-resourced communities may have better data collection infrastructure, leading to more detailed and higher-quality data. In contrast, under-resourced communities may have gaps or inaccuracies, meaning their needs and issues might be underrepresented or misrepresented. This can result in decisions that overlook or fail to address the unique challenges of these communities.

- Algorithmic Bias: If machine learning models are trained predominantly on data from well-resourced areas, they may perform poorly for under-resourced areas. For example, predictive models could be biased towards patterns and behaviors of well-resourced populations, leading to skewed recommendations or predictions that don’t generalize well to diverse populations.

- Resource Allocation: In public policy or social services, data-driven decisions based on biased datasets may disproportionately favor well-resourced communities. This can reinforce existing inequalities, as resources and interventions may be directed toward areas with richer data, potentially leaving under-resourced communities further behind.

To mitigate these biases, it’s crucial to be aware of data imbalances and, where possible, employ techniques such as data re-sampling, careful feature selection, and fairness-aware algorithms to ensure more equitable model outcomes.

**TO DO:** Write 3-5 sentences about TSA.
- Time Series Analysis (TSA) is a method used to analyze data that is collected sequentially over time.
- TSA focuses on identifying patterns within the data, specifically examining trends, seasonality, and residuals.
- The trend component reveals the general direction the data follows over time, such as an upward or downward trajectory.
- Seasonality represents recurring patterns at regular intervals, like increased sales during holidays.
- Lastly, residuals are random fluctuations that do not follow any predictable pattern. By breaking down a time series into these components, TSA helps in making data-driven predictions and uncovering insights about the underlying factors influencing the data.






**Step 2:** Let's begin by creating a chart that displays the average number of crashes per hour of the day. This will help us understand whether additional factors are contributing to crashes - i.e. rush hour, school dismissal time, night fall, etc.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "/content/Motor_Vehicle_Collisions_-_Crashes_20241025.csv"
data = pd.read_csv(file_path, low_memory=False)

# Convert 'CRASH DATE' and 'CRASH TIME' to datetime
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M', errors='coerce')

# Remove rows with invalid 'CRASH TIME'
data = data.dropna(subset=['CRASH TIME'])

# Extract the hour of the day
data['Hour of Day'] = data['CRASH TIME'].dt.hour

# Group by 'Hour of Day' and calculate the number of crashes
crashes_per_hour = data.groupby('Hour of Day').size()

# Plot the average number of crashes
plt.figure(figsize=(12, 6))
sns.barplot(x=crashes_per_hour.index, y=crashes_per_hour.values, palette="viridis",hue=crashes_per_hour)
plt.title('Number of Crashes per Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crashes')
plt.xticks(range(0, 24))
plt.show()
````

**Step 4**: Apply time series decomposition to review trends, seasonality, and residuals. New to time series analysis? Review the [Time Series Flashcard video series](https://youtube.com/playlist?list=PLNs9ZO9jGtUAqd0CNKySksPJJaOqYPMvk&feature=shared) here to learn about trends, components, and further analysis!

````python
!pip install statsmodels
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose

# Count the number of crashes per day, group by CRASH DATE
daily_crashes = data.groupby('CRASH DATE').size()

# Set plot style
sns.set(style="darkgrid")

# Plot the daily crashes time series
plt.figure(figsize=(15, 6))
plt.plot(daily_crashes, label='Daily crashes')
plt.title('Daily Motor Vehicle Collisions in NYC')
plt.xlabel('Date')
plt.ylabel('Number of crashes')
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(daily_crashes, model='additive', period=365)

# Plot the decomposed components
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(15, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonality')
decomposition.resid.plot(ax=ax3)
ax3.set_title('Residuals')
plt.tight_layout()
plt.show()
````

The visualizations above provide valuable insights into the time series of daily motor vehicle collisions in New York City:

1. Time Series Plot: This shows the number of daily crashes over time. You might observe long-term trends, seasonal patterns, or significant outliers.

2. Decomposed Components:
  
    2.1 Trend: This graph shows the long-term trend in the data, which can indicate whether crashes are increasing, decreasing, or stable over time.

    2.2 Seasonality: This reveals any regular patterns that repeat over a specific period, such as yearly. It helps identify times of the year with higher or lower crash frequencies.

    2.3 Residuals: These are the irregular components that cannot be attributed to the trend or seasonality. They might include random or unpredictable fluctuations.

# **Research Question : How do crash frequencies and severities vary across different boroughs during rush hours and non-rush hours?**

- Goal: To analyze the variation in crash patterns across the five boroughs of New York City (Brooklyn, Queens, Manhattan, Bronx, and Staten Island) during rush hours (7–10 AM, 4–7 PM) compared to non-rush hours.



````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the dataset
file_path = "/content/Motor_Vehicle_Collisions_-_Crashes_20241025.csv"
data = pd.read_csv(file_path, low_memory=False)
````

````python
data.columns
````

# - 1. Accidents per Hour Comparison: By comparing the number of accidents during rush and non-rush hours, we can identify time periods with a higher risk of crashes.

````python
data['CRASH DATE'] = pd.to_datetime(data['CRASH DATE'])
data['CRASH TIME'] = pd.to_datetime(data['CRASH TIME'], format='%H:%M', errors='coerce')
data['Hour'] = data['CRASH TIME'].dt.hour
data['Borough'] = data['BOROUGH']

data['Time Period'] = data['Hour'].apply(lambda x: 'Rush Hour' if 7 <= x <= 10 or 16 <= x <= 19 else 'Non-Rush Hour')

boroughs = ['BROOKLYN', 'QUEENS', 'MANHATTAN', 'BRONX', 'STATEN ISLAND']
data = data[data['Borough'].isin(boroughs)]

data['Severity'] = data[['NUMBER OF PERSONS KILLED', 'NUMBER OF PERSONS INJURED']].sum(axis=1).apply(
    lambda x: 'Fatal' if x > 0 else 'Non-Fatal'
)
crash_summary = data.groupby(['Borough', 'Time Period']).size().reset_index(name='Count')

# Calculate accidents per hour
crash_summary['Accidents per Hour'] = crash_summary['Count'] / crash_summary['Time Period'].apply(lambda x: 6 if x == 'Rush Hour' else 18)


sns.set_palette("colorblind")
plt.figure(figsize=(10, 6))
sns.barplot(data=crash_summary, x='Borough', y='Accidents per Hour', hue='Time Period')
plt.title('Accidents per Hour During Rush Hours vs Non-Rush Hours by Borough')
plt.xlabel('Borough')
plt.ylabel('Accidents per Hour')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
crash_summary
````

# 2. Crash Severity Analysis: Assessing how the severity of crashes (fatal vs. non-fatal) varies between boroughs and time periods can help prioritize safety measures in high-risk zones and times.


````python

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette("colorblind")
severity_summary = data.groupby(['Borough', 'Time Period', 'Severity']).size().reset_index(name='Count')
severity_summary['Percentage'] = severity_summary.groupby(['Borough', 'Time Period'])['Count'].transform(lambda x: (x / x.sum()) * 100)

# Plotting the severity trends
plt.figure(figsize=(12, 8))
sns.barplot(data=severity_summary, x='Borough', y='Percentage', hue='Severity', errorbar=None)
plt.title('Percentage Distribution of Crash Severity by Borough and Time Period', fontsize=14)
plt.xlabel('Borough', fontsize=12)
plt.ylabel('Percentage', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Severity', fontsize=10, title_fontsize=12)
plt.tight_layout()
plt.show()
````

# 3. Temporal Trends: Understanding hourly patterns of accidents by borough and time period can help predict when interventions (e.g., traffic controls or emergency response) may be most effective.

````python
#Facet Grid for Hourly Crash Trends by Borough and Time Period ---
hourly_trend = data.groupby(['Hour', 'Borough', 'Time Period']).size().reset_index(name='Count')

# Plotting hourly trends in a facet grid
g = sns.FacetGrid(hourly_trend, col="Borough", hue="Time Period", col_wrap=3, height=4, aspect=1.5)
g.map(sns.lineplot, 'Hour', 'Count', marker='o')
g.set_axis_labels('Hour of the Day', 'Number of Crashes')
g.set_titles("{col_name}")
g.add_legend(title="Time Period")
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Hourly Crash Trends by Borough and Time Period', fontsize=16)
plt.tight_layout()
plt.show()

````

# 4. Statistical Testing (Chi-Square): Running chi-square tests to see if there are statistically significant differences in crash frequencies between rush and non-rush hours can provide solid evidence for targeted interventions.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import chi2_contingency
import numpy as np

data = data.dropna(subset=['Time Period', 'Borough'])

contingency_table = pd.crosstab(data['Time Period'], data['Borough'])

chi2, p, dof, expected = chi2_contingency(contingency_table)
if np.any(expected < 5):
    print("Warning: Some expected frequencies are less than 5. Results may not be reliable.")

plt.figure(figsize=(12, 6))
sns.heatmap(
    contingency_table,
    annot=True,
    fmt='d',
    cmap='Blues',
    cbar=True
)
plt.title('Observed Crash Frequencies by Borough and Time Period')
plt.xlabel('Borough')
plt.ylabel('Time Period')
plt.xticks(rotation=45)
plt.show()

expected_df = pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns)

plt.figure(figsize=(12, 6))
sns.heatmap(
    expected_df,
    annot=True,
    fmt='.2f',
    cmap='Blues',
    cbar=True
)
plt.title('Expected Crash Frequencies by Borough and Time Period')
plt.xlabel('Borough')
plt.ylabel('Time Period')
plt.xticks(rotation=45)
plt.show()

print(f"Chi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4f}")
if p < 0.05:
    print("The differences in crash frequencies across time periods are statistically significant.")
else:
    print("No significant difference in crash frequencies across time periods.")
````

- Conclusion:
Since the P-value is less than 0.05, the differences in crash frequencies across time periods are statistically significant.
This means that time periods do have an impact on the number of crashes, and the observed differences in frequencies are unlikely to be due to chance alone.
